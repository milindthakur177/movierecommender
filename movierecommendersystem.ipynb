```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('movies.csv')
ratings=pd.read_csv('ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
def replace_name(x):
    return movies[movies['movieId']==x].title.values[0]
ratings.movieId= ratings.movieId.map(replace_name)
```

```python
ratings.head()
```

```python
ratings.drop(['timestamp'],axis=1,inplace=True)
```

```python
ratings.head()
```

```python
main=ratings.pivot_table(index=['userId'],columns=['movieId'],values='rating')
```

```python
main.head()
```

```python
main.shape
```

```python
corrmatrix=main.corr(method='pearson', min_periods=25)
```

```python
corrmatrix.head()
```

```python
"""for random user 1"""

myrating=main.loc[1].dropna()
```

```python
myrating.head()
```

```python
simc=pd.Series()
for i in range(0,len(myrating.index)):
    print("Adding"+myrating.index[i])
    sims=corrmatrix[myrating.index[i]].dropna()
    sims=sims.map(lambda x : x*myrating[i])
    simc=simc.append(sims)
    
    
```

```python
print("Sorting the list")
```

```python
simc.sort_values(inplace=True,ascending=False)
```

```python
print(simc.head(10))
```

```python
simc=simc.groupby(simc.index).sum()
```

```python
simc.sort_values(inplace=True,ascending=False)
```
